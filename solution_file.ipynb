```
##Split data into test train
local_train = train_df.loc[train_df.date <= '2020-01-31'].reset_index()
local_test = train_df.loc[train_df.date > '2020-01-31'].reset_index()

print('Local train min date', local_train.date.min())
print('Local train max date', local_train.date.max())

print('Local train length', len(local_train))

print('Local test min date', local_test.date.min())
print('Local test max date', local_test.date.max())

print('Local test length', len(local_test))

```

```
label_encoder = LabelEncoder()
product_categories_df.Category = label_encoder.fit_transform(product_categories_df.Category)
product_categories_df.head()
```

```
##Merge datasets
train_df_merged = pd.merge(local_train,product_categories_df, on='PID')
train_df_merged.head()
```

```
train_date = pd.date_range('2019-11-01','2020-01-30', freq='D')
train_submission_df = pd.DataFrame({'datetime':train_date})
unique_accounts = train_df.acc.unique()


for ac in unique_accounts:
    #train_submission_df[str(ac)] = 0
    transactions = train_df.loc[train_df['acc'] == ac]
    dts = transactions['date'].dt.round('D')
    dates = dts.astype(str).unique()
    train_submission_df.loc[train_submission_df['datetime'].isin(dates), ac] = 1
    
train_submission_df.fillna(0, inplace=True)
train_submission_df.head(10)
```

```
len(train_submission_df[unique_accounts].values.flatten())
```

```
# Reshape this as in sample submission
# I add some extra columns that may be useful

unique_pids = product_categories_df.PID.unique()
train_new = pd.DataFrame({
    'Account X date X PID':np.concatenate([[str(acc) + " X " + str(date).split(" ")[0] + " X " + str(pid) 
                                             for pid in unique_pids 
                                            for date in train_submission_df['datetime']
                                           for acc in unique_accounts]]),
})
train_new.head()
```
